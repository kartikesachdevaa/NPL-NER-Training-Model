```python
import pickle

def load_and_split_skills(pkl_file):
    # Load the list of skill names from the .pkl file
    with open(pkl_file, 'rb') as file:
        skill_list = pickle.load(file)
    
    # Initialize an empty list to store the words
    separated_words = []
    
    # Split each skill name by commas, strip whitespace, and filter out empty strings
    for skill in skill_list:
        words = [word.strip() for word in skill.split(',') if word.strip()]
        separated_words.extend(words)
    
    return separated_words

# Example usage
separated_words = load_and_split_skills('skill_list.pkl')
for word in separated_words:
    print(word)
```

```python
import pandas as pd
import re
from rapidfuzz import process as rf_process, fuzz as rf_fuzz
import numpy as np
import pickle
from tqdm import tqdm
from functools import lru_cache

# Load the search data
df = pd.read_csv('search_clean_data.csv', usecols=['sid', 'query'])

# Load the skill list
with open('skill_list.pkl', 'rb') as file:
    skill_list = pickle.load(file)

# Load the Tags.xlsx and extract the L3 column
tags_df = pd.read_excel('Tags.xlsx', usecols=['L3'])
l3_tags = tags_df['L3'].dropna().unique().tolist()

# Load the pincodes.csv and extract the City column
pincodes_df = pd.read_csv('pincodes.csv', usecols=['City'])
city_list = pincodes_df['City'].dropna().unique().tolist()

# Clean and preprocess the query column
def clean_query(query):
    if pd.isna(query):
        return ""
    query = re.sub(r'\s*,\s*', ',', query.strip().lower())
    query = re.sub(r',+', ',', query)
    query = re.sub(r'[^a-zA-Z0-9, ]', '', query)  # Remove non-alphanumeric characters
    return query

df['cleaned_query'] = df['query'].apply(clean_query)

# Split the cleaned queries into individual entities
def split_query(query):
    if not query:
        return []
    return [entity.strip() for entity in query.split(',') if entity.strip()]

df['entities'] = df['cleaned_query'].apply(split_query)

# Caching the results to avoid redundant calculations
@lru_cache(maxsize=100000)
def cached_tag_entity(entity, skill_list, l3_tags, city_list, skill_threshold=95, jt_threshold=95, city_threshold=95):
    if not entity:
        return 'Not Skill', 'Not JT', 'Not City'
    
    skill_match = rf_process.extractOne(entity, skill_list, scorer=rf_fuzz.WRatio)
    jt_match = rf_process.extractOne(entity, l3_tags, scorer=rf_fuzz.WRatio)
    city_match = rf_process.extractOne(entity, city_list, scorer=rf_fuzz.WRatio)
    is_skill = 'Skill' if skill_match and skill_match[1] >= skill_threshold else 'Not Skill'
    is_jt = 'JT' if jt_match and jt_match[1] >= jt_threshold else 'Not JT'
    is_city = 'City' if city_match and city_match[1] >= city_threshold else 'Not City'
    
    return is_skill, is_jt, is_city

def tag_entities(entities, skill_list, l3_tags, city_list):
    skill_tags = []
    jt_tags = []
    city_tags = []
    for entity in entities:
        is_skill, is_jt, is_city = cached_tag_entity(entity, tuple(skill_list), tuple(l3_tags), tuple(city_list))
        skill_tags.append(is_skill)
        jt_tags.append(is_jt)
        city_tags.append(is_city)
    return skill_tags, jt_tags, city_tags

def process_chunk(chunk, skill_list, l3_tags, city_list):
    skill_tags, jt_tags, city_tags = zip(*chunk['entities'].apply(lambda x: tag_entities(x, skill_list, l3_tags, city_list)))
    chunk['Skill_tag'] = skill_tags
    chunk['JT_tag'] = jt_tags
    chunk['City_tag'] = city_tags
    return chunk

# Chunk the data and process each chunk with tqdm progress bar
chunk_size = 1000
num_chunks = int(np.ceil(len(df) / chunk_size))

processed_chunks = []
for i in tqdm(range(num_chunks), desc="Processing Chunks"):
    start_idx = i * chunk_size
    end_idx = (i + 1) * chunk_size
    chunk = df.iloc[start_idx:end_idx].copy()
    processed_chunk = process_chunk(chunk, skill_list, l3_tags, city_list)
    processed_chunks.append(processed_chunk)

df_processed = pd.concat(processed_chunks)

# Add the L3 column to the processed DataFrame
df_processed['L3'] = df_processed['entities'].apply(lambda x: [entity for entity in x if entity in l3_tags])

# Save the processed DataFrame to a CSV file
df_processed.to_csv('tagged_queries_with_entities.csv', index=False)

# Verify the results
tagged_queries = pd.read_csv('tagged_queries_with_entities.csv')
print("\nFinal DataFrame with Tagged Columns:")
print(tagged_queries.head())
```
